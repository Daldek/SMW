```python
from data_io.excel_reader import ExcelMeasurementReader
from analysis.plots import plot_water_quality_timeseries
```

```python
data = r"data\Wyniki_tabela_wzor NOWY_06-2025.xlsx"
```

```python
def main():
    path = data

    reader = ExcelMeasurementReader()
    points = reader.read_points(path)

    print(f"Załadowano {len(points)} punkt(ów)\n")

    for name, point in points.items():
        print("=" * 60)
        print(f"Kod punktu                    : {point.code}")
        print(f"Współrzędne punktu (lat, lon) : {point.latitude}, {point.longitude}")
        print(f"Nazwa rzeki                   : {point.river_name}")
        print(f"Kod JCWP                      : {point.jcwp_code}")
        print(f"Zarząd Zlewni                 : {point.catchment_authority}")
        print(f"RZGW                          : {point.rzgw}")
        print(f"Nazwa punktu                  : {point.name}")
        print(f"Opis lokalizacji              : {point.location_description}")
        print(f"Otoczenie                     : {point.surroundings}")
        print(f"Osoba badająca                : {point.investigator}")
        print(f"Kontakt                       : {point.contact}")
```

```python
if __name__ == "__main__":
    main()
```
